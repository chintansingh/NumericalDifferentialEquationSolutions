```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

## Known Function Derivative - Analytical Method

$y = f(x) = x^4 $

$y' = f'(x) = 4x^3$

At $x = 3, y' = 4(3^3)$ - This becomes the **slope** of the curve at x =3 <br>

The equation of the tanget line at x = 3 is $y = (slope)x + c$

To make the tangent touch the curve at x = 3, we compute c as:
$c = f(3) - 3*f'(3)$

```python
x = np.linspace(0,5,100)
y = x**4
```

```python
slope = 4*(3**3)

c = 3**4 - 3*slope

y2 = slope*x + c
print(slope)
```

```python
fig = plt.figure(figsize = (16,8))
plt.plot(x,y)
plt.plot(x,y2)
plt.scatter(3,3**4,color = 'r')
```

## Unknown Derivative - Numerical Method

```python
x = np.linspace(0,5,100)
```

```python
def function(x):
    return x**4
```

```python
y = function(x)
```

Now computing slope as $$ slope = \frac{(y_{2} - y_{1})}{(x_{2} - x_{1})} $$

Here step size, $$ h = x_{2} - x_{1} $$

```python
def getSlope(x1,h):
    
    y1 = function(x1)
    x2 = x1 + h
    y2 = function(x2)
    
    return (y2-y1)/(x2-x1)   
```

```python
fig,axes = plt.subplots(2,1, figsize = (10,8))

x1 = 3
h = 1
slope = getSlope(x1,h)
c = function(x1) - slope*x1
y_tangent = slope*x + c

plt.subplot(2,1,1)
plt.plot(x,y)
plt.plot(x,y_tangent)
Xpoints = np.array([x1,x1+h])
Ypoints = slope*Xpoints + c
plt.scatter(Xpoints,Ypoints,color = 'r')
plt.title("h = "+str(h))



x1 = 3
h = 0.1
slope = getSlope(x1,h)
c = function(x1) - slope*x1
y_tangent = slope*x + c

plt.subplot(2,1,2)
plt.plot(x,y)
plt.plot(x,y_tangent)
Xpoints = np.array([x1,x1+h])
Ypoints = slope*Xpoints + c
plt.scatter(Xpoints,Ypoints,color = 'r')
plt.title("h = "+str(h))
```

### Increment Size and Slope accuracy
#### As computed analytically, True Slope is 108.00

```python
print("h","\t","slope")
chk = [1,0.7,0.3,0.1,0.07,0.03,0.01,0.007,0.003,0.001]
slopes = []
for i in chk:
    slopes.append(round(getSlope(x1,i),4))
    print(i,"\t",slopes[-1])

fig = plt.figure(figsize = (16,8))    
plt.plot(chk,slopes)
plt.scatter(chk,slopes)

plt.axhline(108, color="r")

plt.gca().invert_xaxis()
plt.show()
```
